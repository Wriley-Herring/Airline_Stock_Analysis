# Import Historical Data

```python
import yfinance as yf
import matplotlib as plt
import seaborn

DAL = yf.Ticker("DAL") #Delta
LUV = yf.Ticker("LUV") #Southwest
UAL = yf.Ticker("UAL") #United
AAL = yf.Ticker("AAL") #American

hist_DAL = DAL.history(period="90d")
hist_LUV = LUV.history(period="90d")
hist_UAL = UAL.history(period="90d")
hist_AAL = AAL.history(period="90d")
```

# Plot Trends

```python

hist_DAL['Close'].plot(figsize=(16,9))
hist_LUV['Close'].plot(figsize=(16,9))
hist_UAL['Close'].plot(figsize=(16,9))
hist_AAL['Close'].plot(figsize=(16,9))
```
